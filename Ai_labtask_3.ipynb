```
collection = []
choice = 'as your wish'

def greeting():
    print('\nSalamun Alaikum\n')

def line():
    print('~~~~~~~~~~~~~~~~~~')

def show_menu():
    greeting()
    print('MENU')
    line()
    print('1. Add course')
    print('2. Remove course')
    print('3. View course')
    print('4. Exit')
    line()
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        course = input('Which course you want to add? ')
        collection.append(course)
        print('course', course, 'added\n')
    elif choice == '2':
        course = input('Which course you want to remove? ')
        if course in collection:
            collection.remove(course)
            print('Removed course', course)
        else:
            print('Not found', course)
    elif choice == '3':
        print('Show the courses:')
        for course in collection:
            print(course)
    elif choice == '4':
        print('Thank you\n')
    else:
        print('Please enter one of 1 to 4 \n')
```
